```python
import pandas as pd

csv_data = pd.read_csv("crime_report/csv/crime_generation_avg_2021.csv", encoding='cp949')

df = pd.DataFrame(csv_data)

df = df.drop('범죄대분류', axis=1)
df
```

```python
df = df.transpose()
```

```python
df.rename(columns=df.iloc[0], inplace=True)
df = df.drop(df.index[0])

df
```

```python
df = df.dropna(axis=1)

df = df.astype(dtype='int64')

df.dtypes
```

```python
df.to_csv("crime_report/csv/crime_generation_avg_2021.csv")
```
